```python
from scipy import stats
data = df['densité']

mu = data.mean()
sigma = data.std(ddof=1)

# ---- Figure ----
fig, ax = plt.subplots(figsize=(8, 4))

# Histogramme (comme en R : barres remplies, transparence)
n, bins, patches = ax.hist(
    data,
    bins=10,                # à ajuster si besoin
    color="#9999ff",        # bleu clair
    edgecolor="black",
    alpha=0.6
)

# ---- Courbe normale rouge superposée ----
bin_width = bins[1] - bins[0]
x = np.linspace(bins[0], bins[-1], 200)
pdf = stats.norm.pdf(x, mu, sigma)

# On scale la densité pour qu’elle suive la hauteur de l’histogramme
ax.plot(x, pdf * len(data) * bin_width, "r-", linewidth=2)

# ---- Ligne verte = moyenne ----
ax.axvline(mu, color="green", linewidth=2)

# ---- Lignes rouges pointillées = µ ± 3σ (ou 2σ si tu préfères) ----
ax.axvline(mu - 2 * sigma, color="red", linestyle="dashed", linewidth=1.5)
ax.axvline(mu + 2 * sigma, color="red", linestyle="dashed", linewidth=1.5)

# ---- Mise en forme ----
ax.set_title("Histogram")
ax.set_xlabel("densité")
ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# =========================
# 3. Loadings = corrélations (spectres ↔ densité)
# =========================
# Séparer les spectres et la réponse
X_spectra = df.iloc[:, :2594]     # variables spectrales
y = df["densité"]  # remplace par ton nom

# Calcul de la corrélation variable par variable
corr_with_y = X_spectra.corrwith(y)
# On concatène spectres + réponse en un seul DataFrame

plt.figure(figsize=(14,5))
sns.lineplot(x=range(2594), y=corr_with_y.values)
plt.title("Loadings")
plt.xlabel("Longueurs d'ondes (cm-1)")
plt.ylabel("Corrélation")
plt.axhline(0, color="black", linewidth=0.8)
plt.show()
```

```python
# =========================
# 4. Tests de normalité sur 'densité'
# =========================

print("\n=== Tests de normalité sur 'densité' ===")

# Test de Shapiro-Wilk
shapiro_stat, shapiro_p = stats.shapiro(y)
print(f"Shapiro-Wilk   : stat = {shapiro_stat:.4f}, p-value = {shapiro_p:.4g}")

# Test de D'Agostino K²
k2_stat, k2_p = stats.normaltest(y)
print(f"D'Agostino K²  : stat = {k2_stat:.4f}, p-value = {k2_p:.4g}")

# Skewness et kurtosis pour info
skew = stats.skew(y, nan_policy="omit")
kurt = stats.kurtosis(y, nan_policy="omit")
print(f"Asymétrie (skew): {skew:.4f}")
print(f"Curtosis        : {kurt:.4f}")

alpha = 0.05

print("\n=== Interprétation (alpha = 0.05) ===")
if (shapiro_p > alpha) and (k2_p > alpha):
    print(
        "La distribution de 'densité' ne s'écarte pas significativement de la normalité.\n"
        "👉 Tu peux utiliser des tests paramétriques :\n"
        "   - Corrélation de Pearson (densité vs autre variable continue)\n"
        "   - t-test / ANOVA (si comparaison entre groupes)\n"
        "   - Modèles linéaires classiques."
    )
else:
    print(
        "La distribution de 'densité' s'écarte significativement de la normalité.\n"
        "👉 Il vaut mieux privilégier des tests non paramétriques ou transformer la variable :\n"
        "   - Corrélation de Spearman\n"
        "   - Tests de Mann-Whitney / Kruskal-Wallis (comparaison de groupes)\n"
        "   - Transformations (log, Box-Cox, etc.) avant modèles linéaires."
    )
```

### Analyse en composante principale

### Analyse en composante principale (ACP)
Cette anayle va permettre de dégager des tendances ou des regroupements en clusters dans des données de façon non supervisée.\
Afin de mener à bien notre analyse en composé principale nous allons en premier lieu faire ressortir la proportion de variance expliqué en fonction du nombre de composantes principale: à 2 PC on a un cumule de 95.7% de variance,à 8 PC on a un cumule de 99.8% de variance expliqué. Ensuite à l'aide de la méthode du coude nous allons choisir le nombre optimal de composante principale. "2" est le nombre de PC qui permet de cumuler environ 97% de variance. Puis à l'aide de la méthode des K-means nous allons choisir le nombre optimal de cluster (groupe) pouvant exister dans ce jeu de données: il en ressort 4 clusters qu'on arrive à séparer en fonction des deux composantes principales. en fonction de CP 1 on repère 2 groupes d’individus marquants, l’un corrélé négativement, l’autre positivement. En CP2 2 couples d’individus se démarquent également. En combinant les deux composantes principales on voit la scission en 3 groupes avec un petit groupe qui peut également se qui aboutit à 4 groupes. On va maintenant determiner l'origine de ses groupes.

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
spectra_data.shape
```

```python
ACP = PCA(n_components =2)
visual = ACP.fit_transform(spectra_data)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sc2 = axes[1].scatter(visual[:,0], visual[:,0], c=density)
axes[0].set_title("PC1")
sc1 = axes[0].scatter(visual[:,1], visual[:,1], c=density)
axes[1].set_title("PC2")

plt.tight_layout()
plt.show()
```

```python
pca = PCA()
X_pca = pca.fit_transform(spectra_data)
explained_var = pca.explained_variance_ratio_
cum_explained = np.cumsum(explained_var)

# Graphique de la variance expliquée
plt.figure(figsize=(6, 4))
plt.plot(range(1, len(explained_var) + 1), cum_explained, marker="o")
plt.xlabel("Nombre de composantes principales")
plt.ylabel("Variance expliquée cumulée")
plt.title("PCA - Variance expliquée cumulée")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

pca_df = pd.DataFrame(
    X_pca[:, :2],  # PC1 et PC2
    columns=["PC1", "PC2"],
    index=spectra_data.index
)

# =========================================
# 5. Choix du nombre de clusters (méthode du coude)
# =========================================

inertias = []
K_range = range(2, 10)  # tu peux adapter

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init="auto")
    km.fit(pca_df)
    inertias.append(km.inertia_)

plt.figure(figsize=(6, 4))
plt.plot(K_range, inertias, marker="o")
plt.xlabel("Nombre de comp (k)")
plt.ylabel("Inertie (within-cluster sum of squares)")
plt.title("Méthode du coude pour choisir k")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python

# 👉 Choisis k après avoir regardé le graphe du coude
k_opt = 4   # par exemple – à ajuster

kmeans = KMeans(n_clusters=k_opt, random_state=42, n_init="auto")
clusters = kmeans.fit_predict(pca_df)

pca_df["cluster"] = clusters

print("\nRépartition des individus par cluster :")
print(pca_df["cluster"].value_counts())

# =========================================
# 7. Visualisation PC1 vs PC2 coloré par clusters
# =========================================

plt.figure(figsize=(7, 5))
sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="cluster",
    palette="tab10",
    s=60,
    alpha=0.8
)
plt.title(f"PCA (PC1 vs PC2) avec k-means, k={k_opt}")
plt.axhline(0, color="grey", linewidth=0.5)
plt.axvline(0, color="grey", linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
# Récupérer les clusters uniques
cluster_ids = sorted(pca_df["cluster"].unique())

print("\n=== Index des individus par cluster ===\n")

clusters_index_dict = {}  # pour stocker les index cluster par cluster

for cid in cluster_ids:
    indices = pca_df[pca_df["cluster"] == cid].index.tolist()
    clusters_index_dict[cid] = indices
    print(f"Cluster {cid} : {len(indices)} individus")
    print(indices)
    print("-" * 50)
```

#### Clustering
Une fois la visualisation des données en composante principale (PC1 et PC2) terminée, on a identifié les individues de chacun des (4) clusters pour comprendre leurs origines et determiner si cela constitue un élément de corrélation.
Vu qu'on ne dispose d'aucune information supplémentaire hors mis la propriété "densité" qui est déjà corrélé à nos données spectrale on ne peut pas tirer de conclusion hative sur l'origine des cluster naturels formés, on va dans cette optique développer un seul modèle de prédiction pour l'ensemble des données spectrales car aucun des clusters n'a d'origine fiable.

### 1er modèle (sans optimisation)

```python
from sklearn.cross_decomposition import PLSRegression
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
```

```python
model_pls = PLSRegression(n_components=3)
model_svr = SVR(kernel="rbf", C=1.0, epsilon=0.1)

model_pls.fit(X_train, y_train)
model_svr.fit(X_train, y_train)
```

### 2nd modèle avec optimisation par pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, KFold, RandomizedSearchCV
from sklearn.metrics import make_scorer, mean_squared_error
```

```python
param_grid = {
    "pls__n_components": list(range(1, 5)) 
}
model_pls = Pipeline([
    ("scaler", StandardScaler()),
    ("pls", PLSRegression())
])
cv = KFold(n_splits=5, shuffle=True, random_state=42)

grid_pls = RandomizedSearchCV(
    estimator=model_pls,
    param_distributions=param_grid,
    n_iter=20,                                   
    scoring="neg_root_mean_squared_error",       
    cv=cv,
    n_jobs=1,                                    
    random_state=42,
    return_train_score=True,
)

grid_pls.fit(X_train, y_train)

best_model = grid_pls.best_estimator_
best_ncomp = grid_pls.best_params_["pls__n_components"]
best_rmsecv = np.sqrt(-grid_pls.best_score_)

print(f"Meilleur nombre de composantes : {best_ncomp}")
print(f"RMSECV (CV={cv.get_n_splits()} folds) : {best_rmsecv:.4f}")
```

```python
y_pred = best_model.predict(X_test)
```

```python
r2_PLS = r2_score(y_test , y_pred)

rmsec_PLS = np.sqrt(mean_squared_error(y_test , y_pred))

var_pls = explained_variance_score(y_test , y_pred1)


#AFFICHAGE DES RESULTATS ET DU GRAPHE Pred vs Ref
performance_table = pd.DataFrame({
    "Metrique": ['R²' ,'RMSE' , 'Score expliqué'],
    'validation_PLS': [r2_PLS ,rmsec_PLS ,var_pls]    
})
print(performance_table)  
```
